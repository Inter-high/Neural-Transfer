```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from PIL import Image
from torchvision.models import vgg19, VGG19_Weights
```

# 2. model

## 2.1. load pre-trained vgg19

```python
vgg19_model = vgg19(weights=VGG19_Weights.DEFAULT).features.eval()
```

## 2.2. add normalization layer

```python
ImageNet_mean = torch.tensor([0.485, 0.456, 0.406]).to(device)
ImageNet_std = torch.tensor([0.229, 0.224, 0.225]).to(device)

class Normalization(nn.Module):
    def __init__(self, mean, std):
        super(Normalization, self).__init__()
        self.mean = mean.view(-1, 1, 1)
        self.std = std.view(-1, 1, 1)

    def forward(self, img):
        return (img - self.mean) / self.std
```

## 2.3. slice vgg19

```python
model = nn.Sequential(Normalization(ImageNet_mean, ImageNet_std))
i = 0
for layer in vgg19_model.children():
    if isinstance(layer, nn.Conv2d):
        i += 1
        name = f"conv_{i}"
    elif isinstance(layer, nn.ReLU):
        layer = nn.ReLU(inplace=False) # inplace=True <- backprop error
        name = f"relu_{i}"
    elif isinstance(layer, nn.MaxPool2d):
        layer = nn.AvgPool2d(2, 2) # follow the paper
        name = f"pool_{i}"
    elif isinstance(layer, nn.BatchNorm2d):
        name = f"bn_{i}"

    model.add_module(name, layer)

    if name == f'conv_5':
        break

print(model)
```

## 2.4. visualize

```python
def visualize_feature_maps(style_results, content_results, noise_results, 
                           layers=['conv_1', 'conv_2', 'conv_3', 'conv_4', 'conv_5'], title=""):
    fig, axes = plt.subplots(len(layers), 3, figsize=(15, 5 * len(layers)))
    fig.suptitle(title, fontsize=16)

    if len(layers) == 1:
        axes = [axes]

    for i, layer in enumerate(layers):
        # Style feature map
        style_feature_map = style_results[layer].squeeze(0).cpu()
        style_mean_map = torch.mean(style_feature_map, dim=0).detach().numpy()
        axes[i][0].imshow(style_mean_map, cmap="viridis")
        axes[i][0].set_title(f"{layer} (Style)")
        axes[i][0].axis("off")

        # Content feature map
        content_feature_map = content_results[layer].squeeze(0).cpu()
        content_mean_map = torch.mean(content_feature_map, dim=0).detach().numpy()
        axes[i][1].imshow(content_mean_map, cmap="viridis")
        axes[i][1].set_title(f"{layer} (Content)")
        axes[i][1].axis("off")

        # Noise feature map
        noise_feature_map = noise_results[layer].squeeze(0).cpu()
        noise_mean_map = torch.mean(noise_feature_map, dim=0).detach().numpy()
        axes[i][2].imshow(noise_mean_map, cmap="viridis")
        axes[i][2].set_title(f"{layer} (Noise)")
        axes[i][2].axis("off")

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```

```python
model.eval()

style_feature_map = style_tensor.clone().to(device)
content_feature_map = content_tensor.clone().to(device)
noise_feature_map = noise_tensor.clone().to(device)

style_feature_maps = {}
content_feature_maps = {}
noise_feature_maps = {}

for name, layer in model.named_children():
    style_feature_map = layer(style_feature_map)
    style_feature_maps[name] = style_feature_map

    content_feature_map = layer(content_feature_map)
    content_feature_maps[name] = content_feature_map

    noise_feature_map = layer(noise_feature_map)
    noise_feature_maps[name] = noise_feature_map
```

```python
visualize_feature_maps(style_feature_maps, content_feature_maps, noise_feature_maps, title="Before Training: Style vs Content vs Noise Feature Maps")
```

# 3. loss

## 3.1. style-loss

## Gram Matrix 비유: 페인팅

### 페인트의 색상과 질감(채널)
- 화가가 그림을 그릴 때 사용한 다양한 **색상과 질감**이 그림의 스타일을 결정합니다.
- 피처맵의 각 채널은 그림의 특정 **색상이나 질감**을 표현한다고 생각할 수 있습니다.

### 색상 간의 관계
- 스타일은 단순히 어떤 색상이 쓰였는지가 아니라, **색상들 간의 관계와 조화**에서 나옵니다.
- 예를 들어:
  - **파란색과 노란색**이 부드럽게 섞이면 차분한 스타일.
  - **빨간색과 검은색**이 강하게 대조되면 강렬한 스타일.
- Gram Matrix는 이러한 **색상 간의 상관관계**를 수치적으로 캡처합니다.

### 그림의 위치 정보 제거
- 스타일은 그림의 구체적인 **위치**와 상관없이, 색상과 질감의 전반적인 **조화**로 정의됩니다.
- Gram Matrix는 위치 정보(공간적 구조)를 무시하고, 스타일의 본질인 **조화와 패턴**만 남깁니다.


```python
def gram_matrix(feature_map):
    batch, channel, height, width = feature_map.size()

    features = feature_map.view(batch * channel, height * width)
    G = torch.mm(features, features.t())

    return G.div(batch * channel * height * width)
```

```python
def style_loss(input, target):
    input_G = gram_matrix(input)
    target_G = gram_matrix(target)

    loss = F.mse_loss(input_G, target_G)

    return loss
```

## 3.2. content-loss

```python
def content_loss(input, target):
    return F.mse_loss(input, target)
```

# 4. train

## 4.1. training

```python
model.eval()

style_feature_map = style_tensor.clone().to(device)
content_feature_map = content_tensor.clone().to(device)

style_feature_maps = {}
content_feature_maps = {}

for name, layer in model.named_children():
    style_feature_map = layer(style_feature_map)
    style_feature_maps[name] = style_feature_map

    content_feature_map = layer(content_feature_map)
    content_feature_maps[name] = content_feature_map
```

```python
style_layers = ['conv_1', 'conv_2', 'conv_3', 'conv_4', 'conv_5']
content_layers = ['conv_4']


def train(model, style_feature_maps, content_feature_maps, epochs=300, style_weight=1000000, content_weight=1):
    required_noise_feature_map = noise_tensor.clone()
    required_noise_feature_map.requires_grad_(True)

    model.eval()
    model.requires_grad_(False)

    optimizer = optim.LBFGS([required_noise_feature_map])
    epoch = 0

    def closure():
        nonlocal epoch  # epoch을 closure 내부에서 수정 가능하도록 선언
        with torch.no_grad():
            required_noise_feature_map.clamp_(0, 1)
        output = required_noise_feature_map

        optimizer.zero_grad()
        style_losses = 0.0
        content_losses = 0.0
        for name, layer in model.named_children():
            output = layer(output)

            if name in style_layers:
                style_losses += style_loss(output, style_feature_maps[name].detach())
            
            if name in content_layers:
                content_losses += content_loss(output, content_feature_maps[name].detach())
        
        style_losses *= style_weight
        content_losses *= content_weight

        loss = style_losses + content_losses
        loss.backward()

        epoch += 1
        if epoch % 50 == 0:
            print(f"epoch {epoch}/{epochs}:")
            print(f"style-loss: {style_losses.item():.4f} content-loss: {content_losses.item():.4f}")
            print()

        return loss

    while epoch <= epochs:
        optimizer.step(closure)

    with torch.no_grad():
        required_noise_feature_map.clamp_(0, 1)

    return required_noise_feature_map

required_noise_feature_map = train(model, style_feature_maps, content_feature_maps)
```

```python
import numpy as np

styled_image = np.array(untransform(required_noise_feature_map.squeeze(0))).astype(np.uint8)
content_image = np.array(untransform(content_tensor.squeeze(0)))

# styled_image와 content_image의 범위 확인
print("Styled Image Range:", np.min(styled_image), "-", np.max(styled_image))
print("Content Image Range:", np.min(content_image), "-", np.max(content_image))

# 픽셀 값을 0~1로 정규화
styled_image = styled_image / 255.0 if np.max(styled_image) > 1 else styled_image
content_image = content_image / 255.0 if np.max(content_image) > 1 else content_image
```

```python
from skimage.color import rgb2lab, lab2rgb

# Lab 색 공간 변환
styled_lab = rgb2lab(styled_image)
content_lab = rgb2lab(content_image)

# 스타일 이미지의 명도(L) 채널만 사용하고, content 이미지의 a, b 채널 사용
luminance = styled_lab[:, :, 0]  # 명도(L)
content_ab = content_lab[:, :, 1:]  # 색상 정보(a, b)

# 명도와 색상 정보를 결합
combined_lab = np.zeros_like(styled_lab)
combined_lab[:, :, 0] = luminance
combined_lab[:, :, 1:] = content_ab

# 다시 RGB로 변환
final_image = lab2rgb(combined_lab)
```
